```python
import pandas as pd
import os
from itertools import islice, repeat, chain
```

```python
PRList = os.listdir('Summary Physio and EMA/')
PRList
```

```python
ids = [sub[:5] for sub in PRList]
```

```python
ids
```

```python
dfcomp = pd.DataFrame()
for PR in ids:
    curdf= pd.read_csv('Summary Physio and EMA/' + PR + ' Survey Window Summary.csv') 
    curdf['PR'] = PR
    print(curdf.size)
    if dfcomp.empty:
        dfcomp = curdf
    else:
        dfcomp = pd.concat([dfcomp,curdf])
print(dfcomp.size)
```

```python
dfcomp.to_csv('Full Summary Features.csv', index=False)
```
